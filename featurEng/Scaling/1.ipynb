```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import seaborn as sns
from sklearn.preprocessing import StandardScaler

from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()

data = pd.DataFrame(data.data,data.target,columns = data.feature_names)
data.head()
```

```python
data.describe()
```

```python
sc = StandardScaler().set_output(transform="pandas")

dft = sc.fit_transform(data)
np.round(dft.describe(),1)
```

```python
dft.hist(bins=50,color='orange',figsize=(12,6))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().set_output(transform="pandas")

dft = scaler.fit_transform(data)
np.round(dft.describe(),1)
```

```python
dft.hist(bins=50,color='orange',figsize=(12,6))
```

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler().set_output(transform="pandas")

dft = sc.fit_transform(data)
np.round(dft.describe(),1)
```

```python
dft.hist(bins=50,color='orange',figsize=(12,6))
```

```python
from sklearn.preprocessing import MaxAbsScaler
sc = MaxAbsScaler().set_output(transform="pandas")

dft = sc.fit_transform(data)
np.round(dft.describe(),1)
```

```python
dft.hist(bins=50,color='orange',figsize=(12,6))
```
